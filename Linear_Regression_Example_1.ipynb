```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.2f}'.format

%matplotlib inline
```

**Load the Dataset**

```python
df = pd.read_csv('USA_Housing.csv')
df
```

**Separate Features and Target Variable**

```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y = df['Price']
```

```python
x
```

```python
y
```

```python
# ML related imports
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

**Split into Train and Test sets**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**Train the Linear Regression Model**

```python
model = LinearRegression() # Untrained model (Not yet fitted to the data)
model
```

```python
model.fit(x_train, y_train) # Train the model i.e. fit the model to the training data (Learn the coefficients)
```

```python
print(model.intercept_)
```

```python
feature_importance = [float(f'{value:f}') for value in model.coef_]
print('Coefficients: \n', feature_importance)
```

```python
# Feature importance in a more readable format
feature_importance_df = pd.DataFrame(data=feature_importance, index=x.columns, columns=['Feature Importance'])
feature_importance_df.sort_values(by='Feature Importance', ascending=False)
```

```python
feature_importance_df.sort_values(by='Feature Importance', ascending=False).plot(
    kind='bar', figsize=(8, 6), title='Feature Importance', legend=False, ylabel='Coefficient Value', xlabel='Feature', color='salmon')

plt.xticks(rotation=75)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Evaluate the Model**

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print('mean_squared_error: ', mean_squared_error(y_test, y_pred))
```

```python
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
plt.scatter(y_test, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('Housing_Model.png')
plt.show()
```

**Make a prediction**

```python
x_new = pd.DataFrame([
    [58204.22309, 6.301814027, 7.375616033, 5.41, 46663.40274],
    [62085.2764, 5.739410844, 7.091808104, 5.49, 44922.1067]
], columns=x.columns)

actual_price = [1059033.312, 1030591.429]
predicted_price = model.predict(x_new)

for i in range(len(x_new)):
    print(f'Actual Price: ${actual_price[i]:,.2f}')
    print(f'Predicted Price: ${predicted_price[i]:,.2f}')
    print('---' * 10)
```

**Save the Model**

```python
file_name = 'housing_model.pkl'
joblib.dump(model, file_name) # Serialize the model and save it to a file
print(f'Model saved to {file_name}')
```

**Load the Saved Model**

```python
file_name = 'housing_model.pkl'
loaded_model = joblib.load(file_name) # Deserialize the model from the file
print(f'Model loaded from {file_name}')
```
