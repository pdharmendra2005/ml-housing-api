```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.2f}'.format

%matplotlib inline
```

**Load the Dataset**

```python
df = pd.read_csv('USA_Housing.csv')
df
```

**Separate Features and Target Variable**

```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y = df['Price']
```

```python
x
```

```python
y
```

```python
# ML related imports
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

**Split into Train and Test sets**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**Train the Linear Regression Model**

```python
model = LinearRegression() # Untrained model (Not yet fitted to the data)
model
```

```python
model.fit(x_train, y_train) # Train the model i.e. fit the model to the training data (Learn the coefficients)
```

```python
print(model.intercept_)
```

```python
feature_importance = [float(f'{value:f}') for value in model.coef_]
print('Coefficients: \n', feature_importance)
```

```python
# Feature importance in a more readable format
feature_importance_df = pd.DataFrame(data=feature_importance, index=x.columns, columns=['Feature Importance'])
feature_importance_df.sort_values(by='Feature Importance', ascending=False)
```

```python
feature_importance_df.sort_values(by='Feature Importance', ascending=False).plot(
    kind='bar', figsize=(8, 6), title='Feature Importance', legend=False, ylabel='Coefficient Value', xlabel='Feature', color='salmon')

plt.xticks(rotation=75)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Evaluate the Model**

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print('mean_squared_error: ', mean_squared_error(y_test, y_pred))
```

```python
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
plt.scatter(y_test, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('Housing_Model.png')
plt.show()
```

**Hyperparameter Tuning**

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV
```

```python
import multiprocessing
multiprocessing.cpu_count()
```

```python
# Using GridSearchCV to find the best alpha for Ridge Regression
param_grid = {
    'alpha': [0.01, 0.1, 0.2, 0.5, 0.8, 1.0, 10.0, 100.0, 1000.0]
}

ridge = Ridge()

grid_search_ridge = GridSearchCV(estimator=ridge, param_grid=param_grid, cv=5, scoring='r2',  n_jobs=-1)
grid_search_ridge.fit(x_train, y_train)

best_ridge_model = grid_search_ridge.best_estimator_
best_r2_score = grid_search_ridge.best_score_
print(f'Best Alpha for Ridge Regression: {best_ridge_model.alpha}')
print(f'Best R2 Score from Grid Search: {best_r2_score:.4f}')

feature_importance = [float(f'{value:f}') for value in best_ridge_model.coef_]
feature_importance_df = pd.DataFrame(data=feature_importance, index=x.columns, columns=['Feature Importance'])
feature_importance_df.sort_values(by='Feature Importance', ascending=False)
```

```python
# Using GridSearchCV to find the best alpha for Lasso Regression
lasso = Lasso()
grid_search_lasso = GridSearchCV(estimator=lasso, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search_lasso.fit(x_train, y_train)
best_lasso_model = grid_search_lasso.best_estimator_
best_r2_score_lasso = grid_search_lasso.best_score_
print(f'Best Alpha for Lasso Regression: {best_lasso_model.alpha}')
print(f'Best R2 Score from Grid Search: {best_r2_score_lasso:.4f}')

feature_importance = [float(f'{value:f}') for value in best_lasso_model.coef_]
feature_importance_df = pd.DataFrame(data=feature_importance, index=x.columns, columns=['Feature Importance'])
feature_importance_df.sort_values(by='Feature Importance', ascending=False)
```

**Hyperpatemeter Tuning with Normalization**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipeline1 = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

pipeline1.fit(x_train, y_train)
y_pred_pipeline1 = pipeline1.predict(x_test)
print('R2 Score for Pipeline with Linear Regression: ', r2_score(y_test, y_pred_pipeline1))
```

```python
param_grid = {
    'alpha': [0.01, 0.1, 0.2, 0.5, 0.8, 1.0, 10.0, 100.0, 1000.0]
}
pipeline_ridge = Pipeline([
    ('scaler', StandardScaler()),
    ('model', Ridge())
])
grid_search_ridge_pipeline = GridSearchCV(estimator=pipeline_ridge, param_grid={'model__alpha': param_grid['alpha']}, cv=5, scoring='r2', n_jobs=-1)
grid_search_ridge_pipeline.fit(x_train, y_train)
best_ridge_pipeline = grid_search_ridge_pipeline.best_estimator_
best_ridge_pipeline_alpha = grid_search_ridge_pipeline.best_params_['model__alpha']
print(f'Best Alpha for Ridge Regression Pipeline: {best_ridge_pipeline_alpha}')
y_pred_ridge_pipeline = best_ridge_pipeline.predict(x_test)
print('R2 Score for Ridge Regression Pipeline: ', r2_score(y_test, y_pred_ridge_pipeline))
```

```python
pipeline_lasso = Pipeline([
    ('scaler', StandardScaler()),
    ('model', Lasso())
])
grid_search_lasso_pipeline = GridSearchCV(estimator=pipeline_lasso, param_grid={'model__alpha': param_grid['alpha']}, cv=5, scoring='r2', n_jobs=-1)
grid_search_lasso_pipeline.fit(x_train, y_train)
best_lasso_pipeline = grid_search_lasso_pipeline.best_estimator_
best_lasso_pipeline_alpha = grid_search_lasso_pipeline.best_params_['model__alpha']
print(f'Best Alpha for Lasso Regression Pipeline: {best_lasso_pipeline_alpha}')
y_pred_lasso_pipeline = best_lasso_pipeline.predict(x_test)
print('R2 Score for Lasso Regression Pipeline: ', r2_score(y_test, y_pred_lasso_pipeline))
```

**Mixing L1 and L2 Regularization in Training activity**

```python
param_grid = [
    {
        'model' : [Lasso()],
        'model__alpha': [0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
    },
    {
        'model' : [Ridge()],
        'model__alpha': [0.01, 0.1, 0.2, 0.5, 0.8, 1.0, 10.0, 100.0, 1000.0],
    }
]

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', None)  # Placeholder, will be set by GridSearchCV
])

grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=20, scoring='r2', n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)
best_model = grid_search.best_estimator_
best_model_type = type(best_model.named_steps['model']).__name__
best_model_alpha = best_model.named_steps['model'].alpha
print(f'Best Model: {best_model_type} with Alpha: {best_model_alpha}')
y_pred = best_model.predict(x_test)
print('R2 Score for Best Model Pipeline: ', r2_score(y_test, y_pred))
```
